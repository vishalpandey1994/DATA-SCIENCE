```python
import numpy as np
import pandas as pd
import math
```

# Importing the Dataset and rounding off to two decimal places


```python
data=pd.read_csv("weight-height.csv").round(2)
```

# Splitting the data into male and female


```python
Mmale=data[data.Gender=="Male"]
Ffemale=data[data.Gender=="Female"]
```

# Removing Index Column from both the datasets

```python
#Mmale=Mmale.drop("Index",axis=1,inplace=False)
#Ffemale=Ffemale.drop("Index",axis=1,inplace=False)
```

# Splitting the data into dependent and independent variables

```python
Male=Mmale.groupby("Height").mean().astype(int)
Female=Ffemale.groupby("Height").mean().astype(int)
```

  # Reseting Index Using reset_index() function

```python
Male=Male.reset_index()
Female=Female.reset_index()
```

```python
Male.info()
```

# Splitting the data into height and weight

```python
X_Male=Male.iloc[:,:1].values
y_Male=Male.iloc[:,1:].values
X_Female=Female.iloc[:,:1].values
y_Female=Female.iloc[:,1:].values
```

# Splitting the data into training and testing datasets

```python
from sklearn.model_selection import train_test_split
```

```python
X_Male_train,X_Male_test,y_Male_train,y_Male_test=train_test_split(X_Male,y_Male,test_size=0.05,random_state=4)
X_Female_train,X_Female_test=train_test_split(X_Female,test_size=0.05,random_state=4)
y_Female_train,y_Female_test=train_test_split(y_Female,test_size=0.05,random_state=4)
```

```python
def predict(Gender):
        from sklearn.linear_model import LinearRegression
        male_regressor=LinearRegression()
        female_regressor=LinearRegression()
        if Gender=="Male":
            male_regressor.fit(X_Male_train,y_Male_train)
            hhh=male_regressor.predict(X_Male_test)
            return hhh
        elif Gender=="Female":
            female_regressor.fit(X_Female_train,y_Female_train)
            gg=female_regressor.predict(X_Female_test)
            return gg
```

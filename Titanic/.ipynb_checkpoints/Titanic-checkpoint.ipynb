# Adding Train and Test Data.

```python
train["Type"]='Train'
test["Type"]='Test'
```

```python
full=pd.concat([train,test])
```

It Seems that only "Age" is Having the Null Values.Thus replacing it by its mean values.

```python
full.isnull().sum()
```

```python
#Replacing the Null Values in the Age with the mean values.
full["Age"].fillna(full["Age"].mean(),inplace=True)
```

```python
full.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
full["Sex"]=encoder.fit_transform(full["Sex"])
```

# Splitting into training and testing data as required.

```python
train_modified=full[full["Type"]=="Train"]
test_modified=full[full["Type"]=="Test"]
```

# Splitting the training data into dependent and independent variable


```python
train_modified
```

```python
full.dtypes
```

```python
X=train_modified.drop(labels=["Cabin","Fare","Name","Parch","SibSp","Ticket","Survived","PassengerId","Embarked","Type"],axis=1)
y=train_modified["Survived"]
X["Age"]=X["Age"].astype(int)
y=y.astype(int)
```

```python
X_test=test_modified.drop(labels=["Cabin","Fare","Name","Parch","SibSp","Survived","Ticket","PassengerId","Embarked","Type"],axis=1)
X_test["Age"]=X_test["Age"].astype(int)
```

```python
from sklearn.linear_model import LogisticRegression
regressor=LogisticRegression()
```

```python
regressor.fit(X,y)
```

```python
y_test=regressor.predict(X_test)
```

```python
y_test
```

```python
test["Survived"]=y_test
```

```python
pd.crosstab(test["Sex"],test["Survived"],margins=True)
```

```python
pd.crosstab(train["Sex"],train["Survived"],margins=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
cl=DecisionTreeClassifier()
```

```python
cl.fit(X,y)
```

```python
Ytest=cl.predict(X_test)
```

```python
test["ssuurvived"]=Ytest
```

```python
pd.crosstab(test["Sex"],test["ssuurvived"],margins=True)
```
